```python
import pandas as pd
from scipy.io import arff
```

```python
# ARFF 파일 경로
arff_file_path = "/Users/taeberie/Desktop/Capstone25SS/ISCX-VPN-NonVPN-2016 CSVs/Scenario A1-ARFF/TimeBasedFeatures-Dataset-60s-VPN.arff"
# 저장할 CSV 파일 경로
csv_file_path = "/Users/taeberie/Desktop/Capstone25SS/A1_60s.csv"

# ARFF 파일 읽기
data, meta = arff.loadarff(arff_file_path)

# pandas DataFrame으로 변환
df = pd.DataFrame(data)

# 만약 문자열이 byte로 저장된 경우 디코딩 필요
for col in df.select_dtypes([object]):
    df[col] = df[col].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)

# CSV로 저장
df.to_csv(csv_file_path, index=False)

print("변환 완료! CSV 파일이 저장되었습니다.")
```
